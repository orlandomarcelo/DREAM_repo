```python
def get_PAR(AL):
    popt_quad = [  0.1628436 ,   2.75648771, -17.10740481]
    popt_lin = [  7.09300446, -50.12770528]
    if AL > 15:
        return popt_lin[0]*AL - popt_lin[1]
    elif AL <= 4.5:
        return 0
    else:
        return popt_quad[0]*AL**2 + popt_quad[1]*AL + popt_quad[2]
    

```

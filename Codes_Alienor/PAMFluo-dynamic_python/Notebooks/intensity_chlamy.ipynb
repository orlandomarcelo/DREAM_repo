```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import alienlab.plot
from mvgavg import mvgavg
p = alienlab.plot.ShowFigure()
g = alienlab.plot.PlotFigure()
%matplotlib inline

def averaging(measured_signal, 
                time_array, window, 
                do_binning = True):

    average_output = mvgavg(measured_signal, window, axis = 1, binning = do_binning)
    downscaled_time = time_array[window//2 :: window] 
    #todo: if binning == False, time array
    return average_output, downscaled_time    

video_set = {}
intensity_set = {}
fluorescence_set = {}
```

```python
intensity_set["DO1"] = pd.read_csv("../2021-01-28_15_37_intensity_DO1/2021-01-28_15_38_intensity_range.csv")
#intensity_set["DO0"] = pd.read_csv("../2021-01-28_14_29_intensity_response_D0/2021-01-28_14_30_intensity_range.csv")
intensity_set["DO2"] = pd.read_csv("../2021-01-28_15_40_intensity_DO2/2021-01-28_15_41_intensity_range.csv")
intensity_set["DO13"] = pd.read_csv("../2021-01-28_16_05_intensity_DO13/2021-01-28_16_26_intensity_range.csv")

fluorescence_set["DO1"] = pd.read_csv("../2021-01-28_15_37_intensity_DO1/2021-01-28_15_38_fluo_range.csv")
#fluorescence_set["DO0"] = pd.read_csv("../2021-01-28_14_29_intensity_response_D0/2021-01-28_14_30_fluo_range.csv")
fluorescence_set["DO2"] = pd.read_csv("../2021-01-28_15_40_intensity_DO2/2021-01-28_15_41_fluo_range.csv")
fluorescence_set["DO13"] = pd.read_csv("../2021-01-28_16_05_intensity_DO13/2021-01-28_16_26_fluo_range.csv")
```

```python
import serial
import math
import time
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget

# Define the frequency and amplitude of the sinus function
freq = 1  # in Hz
ampmax = 100
amp_factor = 0.2
offset_factor = 0.1

# Define the sinus function to generate the PWM value for the LED
def sin_wave(t,freq,amp):
    return int(amp * math.sin(2 * math.pi * freq * t) + amp)


t = np.linspace(0,10,1000)
signal = np.round(amp_factor*ampmax*np.sin(2* np.pi* t * freq) + (offset_factor)*ampmax, 0)
```

```python
plt.plot(t,signal)
```
